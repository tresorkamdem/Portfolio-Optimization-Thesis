# Datenaufbereitung für Portfoliooptimierung

Dieses Notebook zeigt die Transformation von Preisen zu Renditen sowie die Berechnung grundlegender statistischer Größen.

## Finanzmathematischer Hintergrund

In der Portfolio-Theorie werden nicht die Preise direkt verwendet, sondern deren Veränderungen.
Diese Veränderungen werden als Renditen bezeichnet.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

## Simulation von Renditen

Wir erzeugen künstliche Renditen, um die grundlegenden Konzepte zu verstehen.

```python
np.random.seed(42)
```

```python
returns = pd.DataFrame({
    "Asset_A": np.random.normal(0.001, 0.02, 100),
    "Asset_B": np.random.normal(0.001, 0.015, 100)
})
```

## Konstruktion von Preisen

Aus Renditen können Preise durch kumulative Multiplikation berechnet werden.

```python
prices = (1 + returns).cumprod()
prices.head()
prices.plot(title="Simulierte Preise")
plt.show()
```

## Logarithmische Renditen

Die Renditen werden berechnet als:

r_t = log(P_t / P_{t-1})

```python
log_returns = np.log(prices / prices.shift(1)).dropna()
log_returns.head()
```

## Statistische Kennzahlen

Wir berechnen:

- μ: durchschnittliche Rendite
- Σ: Kovarianzmatrix

```python
mu = log_returns.mean()
sigma = log_returns.cov()

mu, sigma
```
