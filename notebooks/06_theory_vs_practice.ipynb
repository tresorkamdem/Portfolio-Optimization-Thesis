## 1. Einführung

In der Finanztheorie spielt die Portfoliooptimierung eine zentrale Rolle bei der Entscheidungsfindung von Investoren.

Klassische Modelle wie die Markowitz-Theorie basieren auf der quantitativen Analyse von Renditen und Risiken. Ziel ist es, ein Portfolio so zu konstruieren, dass entweder die Rendite maximiert oder das Risiko minimiert wird.

Im Rahmen dieser Arbeit wird untersucht, inwiefern moderne datengetriebene Methoden, insbesondere Machine Learning, die klassische Portfoliooptimierung erweitern können.

Zur besseren Einordnung der theoretischen Ergebnisse wird zusätzlich ein vereinfachtes praxisbasiertes Portfolio betrachtet.

---

## 2. Theoretische Grundlagen

Die moderne Portfoliotheorie nach Markowitz (1952) basiert auf der Annahme eines Trade-offs zwischen Risiko und Rendite.

Zentrale Konzepte sind:

- Erwartete Rendite eines Portfolios  
- Varianz bzw. Standardabweichung als Risikomaß  
- Kovarianz zwischen den Assets  
- Diversifikation zur Risikoreduktion  

Mathematisch wird das Risiko eines Portfolios wie folgt beschrieben:

σ² = wᵀ Σ w  

Dabei bezeichnet:
- w den Vektor der Portfolio-Gewichte  
- Σ die Kovarianzmatrix der Renditen  

Ein effizientes Portfolio ist dadurch gekennzeichnet, dass es für ein gegebenes Risiko die maximale Rendite erzielt.

---

## 3. Praxisbezug (Einordnung)

In der Praxis treffen Investoren ihre Entscheidungen häufig nicht ausschließlich auf Basis mathematischer Modelle.

Stattdessen spielen folgende Faktoren eine Rolle:

- Marktbeobachtung  
- institutionelle Rahmenbedingungen  
- individuelle Risikobereitschaft  
- Erfahrungswerte  

Ein vereinfachtes praxisbasiertes Portfolio kann daher als Referenz dienen, um die Ergebnisse der quantitativen Modelle besser einzuordnen.

Es ist wichtig zu betonen, dass ein solches Portfolio keine Optimierung im mathematischen Sinne darstellt, sondern lediglich eine illustrative Vergleichsbasis bildet.

## Fazit

Die Portfoliooptimierung stellt ein zentrales Problem der Finanzmathematik dar.

Während theoretische Modelle eine klare Struktur zur Bewertung von Risiko und Rendite liefern, zeigt die Praxis, dass Investitionsentscheidungen häufig auch von nicht-quantitativen Faktoren beeinflusst werden.

Die Kombination beider Perspektiven ermöglicht ein umfassenderes Verständnis von Portfolioentscheidungen und bildet die Grundlage für die weiteren Analysen in diesem Projekt.
