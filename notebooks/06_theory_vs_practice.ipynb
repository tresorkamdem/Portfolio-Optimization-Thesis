# 🇩🇪 1. Einführung

In der Finanztheorie spielt Risiko eine zentrale Rolle bei Investitionsentscheidungen. Klassische Modelle wie das Markowitz-Modell beschreiben Risiko explizit mithilfe statistischer Größen wie Varianz und Kovarianz der Renditen.

In der Praxis wird Risiko jedoch häufig anders wahrgenommen. Insbesondere bei Staatsanleihen oder stark regulierten Finanzmärkten wird das Risiko durch institutionelle Mechanismen reduziert, sodass es von Investoren oft als sehr gering oder nahezu nicht existent angesehen wird.

Ziel dieses Notebooks ist es, die Differenz zwischen:

- quantitativer Risikomodellierung (Theorie)
- institutioneller Risikowahrnehmung (Praxis)

zu analysieren.

---

# 🇬🇧 1. Introduction

In financial theory, risk plays a central role in investment decisions. Classical models such as the Markowitz framework explicitly model risk using statistical measures like variance and covariance of asset returns.

In practice, however, risk is often perceived differently. Especially in sovereign bond markets, institutional mechanisms reduce risk to a level that is often considered negligible.

---

# 🇩🇪 2. Theoretische Perspektive: Risiko-Rendite-Abwägung

Die moderne Portfoliotheorie (Markowitz, 1952) basiert auf der Annahme, dass Investoren einen Trade-off zwischen Risiko und erwarteter Rendite eingehen.

Wichtige Punkte:

- Risiko wird durch Varianz oder Standardabweichung gemessen  
- Diversifikation reduziert das Risiko eines Portfolios  
- Ein vollständig risikofreies Portfolio existiert in der Realität nicht  

Formal gilt:

> Risiko > 0

Selbst optimal konstruierte Portfolios enthalten Unsicherheit.

---

# 🇬🇧 2. Theoretical Perspective

Modern Portfolio Theory assumes a trade-off between risk and expected return. Even optimal portfolios still carry uncertainty.

---

# 🇩🇪 3. Praktische Perspektive: Institutionelle Absicherung

In realen Finanzmärkten, insbesondere im Bereich von Staatsanleihen (z. B. CEMAC-Region), wird Risiko häufig als sehr gering wahrgenommen.

Institutionelle Mechanismen umfassen:

- Zentralbank (z. B. BEAC)  
- Automatische Belastungsvereinbarungen zwischen Staaten  
- Treuhandkonten (séquestre)  
- Regulatorische Sicherungssysteme  

Diese Mechanismen haben das Ziel:

- die Rückzahlung an Investoren sicherzustellen  
- die Ausfallwahrscheinlichkeit zu minimieren  
- Vertrauen in den Markt zu schaffen  

In der Praxis führt dies oft zu folgender Wahrnehmung:

> „Das Risiko ist sehr gering oder praktisch nicht vorhanden.“

---

# 🇬🇧 3. Practical Perspective

In practice, institutional frameworks (central banks, guarantees, escrow accounts) reduce perceived risk significantly.

---

# 🇩🇪 4. Kritische Analyse

Trotz dieser institutionellen Sicherungsmechanismen kann Risiko nicht vollständig eliminiert werden.

Zu berücksichtigen sind:

- Staatsinsolvenzen sind selten, aber möglich  
- Makroökonomische Schocks können Staaten beeinflussen  
- Politische Risiken können die Rückzahlung gefährden  
- Historische Beispiele zeigen, dass auch Staaten ausfallen können  

Daher gilt:

> Risiko wird nicht eliminiert, sondern reduziert, verteilt und kontrolliert.

---

# 🇬🇧 4. Critical Analysis

Even strong institutional guarantees cannot eliminate risk completely. Risk is reduced and redistributed, but not removed.

---

# 🇩🇪 5. Zentrale Erkenntnis

Es besteht eine wesentliche Differenz zwischen:

- modelliertem Risiko (quantitative Modelle)  
- wahrgenommenem Risiko (Praxis und Institutionen)  

Diese Differenz kann wie folgt beschrieben werden:

> Diskrepanz zwischen theoretischem Risiko und praktischer Risikowahrnehmung

---

# 🇬🇧 5. Key Insight

There exists a gap between modeled risk and perceived risk.

---

# 🇩🇪 6. Fazit

Dieses Projekt zeigt, dass:

- die Finanztheorie Risiko quantitativ misst

Das Verständnis dieser Differenz ist entscheidend für die Anwendung von Portfoliooptimierung in realen Märkten.
